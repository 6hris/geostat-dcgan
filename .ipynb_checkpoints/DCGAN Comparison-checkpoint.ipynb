```python
from PIL import Image
import os
import pandas as pd
import numpy as np
import skgstat as sgs
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import random
```

```python
sgs_df = pd.read_parquet("GL_small_ensemble.parquet.gzip", engine = "fastparquet")
```

```python
coords = sgs_df[['X','Y']].values.reshape(100,100,2)
coords = coords[:96,:96].reshape(9216,2)
sgs_val = sgs_df['sim_2'].values.reshape(100,100)[:96,:96].reshape(9216)
```

```python
sgs_val = (sgs_val+1)/2
```

```python
def calculations(image_folder):
    # calculate mean and variance
    
    results = np.empty((len(os.listdir(image_folder)), 96, 96))
    
    for i, name in enumerate(os.listdir(image_folder)):
        path = os.path.join(image_folder, name)
        image = Image.open(path).convert('L')
        image_normalized = np.array(image, dtype=np.float32) / 255.0
        
        
        results[i] = image_normalized

    return results
```

```python
image_folder = 'individual_images'
nImages = len(os.listdir(image_folder))
results = calculations(image_folder)
```

```python
def plot_colormap(X, Y, Color, colorbar_title, title):

    fig = plt.figure(figsize = (6,5))
    ax = plt.gca()
    im = ax.scatter(X, Y, c=Color, marker='.', s=5, cmap='gist_earth')
    plt.title(title)
    plt.xlabel('X [m]'); plt.ylabel('Y [m]')
    plt.locator_params(nbins=5)
    plt.axis('scaled')

    # make colorbar
    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.1)
    cbar = plt.colorbar(im, ticks=np.linspace(np.min(Color), np.max(Color), 11), cax=cax)
    cbar.set_label(colorbar_title, rotation=270, labelpad=15)
   
```

```python
plot_colormap(coords[:,0], coords[:,1], np.mean(results, axis = 0), "A", "B")
```

```python
plot_colormap(coords[:,0], coords[:,1], np.var(results, axis = 0), "A", "B")
```

```python
random.seed(0)
samples = random.sample(range(nImages), 15)

maxlag = 50000      # maximum range distance
n_lags = 70         # num of bins
```

```python
SGS_var = sgs.Variogram(coords, sgs_val, bin_func='even', n_lags=n_lags, maxlag=maxlag, normalize=False)
```

```python
sampled_varios = []
for i in samples:
    vario = sgs.Variogram(coords, results[i].reshape(9216), bin_func='even', n_lags=n_lags, maxlag=maxlag, normalize=False)
    variodat = vario.experimental
    sampled_varios.append(variodat)
```

```python
# V1 = sgs.Variogram(coords, results[0].reshape(9216), bin_func='even', n_lags=n_lags, maxlag=maxlag, normalize=False)
```

```python
xdata = SGS_var.bins
```

```python
from itertools import cycle
cycol = cycle('bgrcm')
```

```python
SGSdata = SGS_var.experimental
```

```python
fig = plt.figure(figsize=(6,4))
plt.scatter(xdata, SGSdata, s=12, c='k', marker='*', label='SGS Sim')

for data in sampled_varios:
    plt.scatter(xdata, data, s=12, c= next(cycol), marker='*')
plt.title(f'Isotropic Experimental Variogram')
plt.legend(loc="upper left")
plt.xlabel('Lag (m)'); plt.ylabel('Semivariance')
fig.savefig(f'vario_sgs_dcgan.png', dpi=fig.dpi)
```
